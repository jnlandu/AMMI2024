### Kernel function definition
A function $k: \mathcal{X} \times \mathcal{X} \to \mathbb{R}$ is a **positive definite (p.d.) kernel** if and only if
$\forall n$, $\forall a \in \mathbb{R}^n$, $x_i \in \mathcal{X}$, 
$$\sum_{i,j = 1}^{n} a_i a_j k(x_i, x_j) \geq 0$$

In matrix notation, this condition is:

$\forall n$, $\forall \mathbf{a} \in \mathbb{R}^n$, $x_i \in \mathcal{X}$,
$$\mathbf{a}^\top K \mathbf{a} \geq 0 $$
where $K$ is the $(n, n)$ Gram matrix of $x$: $K_{i,j} = k(x_i, x_j)$

---

It is also equivalent to $K$ being **positive semi-definite** (symmetric and all its eigenvalues are $\geq 0$ )

### Exercise 1

Let $k_1$ and $k_2$ be p.d. kernels.

In each of the following cases, **show that $k$ is a p.d. kernel**

**(a)** $k(x, y) = \gamma k_1(x, y)$ with $\gamma > 0$
- $k$ is symmetric
- Let $n\geq 1$, let $x_1,...,x_n \in \mathcal{X}$ and let $K = [k(x_i, x_jà]_{i,j=1,...,n}$ and $K_1 = [k_1(x_i, x_jà]_{i,j=1,...,n}$. <br>
$K$ is p.s.d., since for any $\alpha \in \mathbb{R}^n$, <br>
 $$ \alpha^T (\gamma K_1) \alpha = \gamma (\alpha^T K_1 \alpha ) \geq 0 $$

**(b)** $k = k_1 + k_2$
- $k$ is symmetric
- Let $n\geq 1$, let $x_1,...,x_n \in \mathcal{X}$ and let $K = [k(x_i, x_j]_{i,j=1,...,n}$, $K_1 = [k_1(x_i, x_j]_{i,j=1,...,n}$ and $K_2 = [k_2(x_i, x_jà]_{i,j=1,...,n}$. <br>
$K$ is p.s.d., since for any $\alpha \in \mathbb{R}^n$, <br>
 $$ \alpha^T (K_1 + K_2) \alpha =  \alpha^T K_1 \alpha  + \alpha^T K_2 \alpha \geq 0 $$

**(c)** Let $f$ be a function $f: \mathcal{X} \to \mathbb{R}$:
$$k(x, y) = f(x)\, k_1(x, y)\, f(y)$$

Let $n \in \mathbb{N}, x_1,...,x_n \in \mathcal{X}, \alpha \in \mathbb{R}^n$. 
$$
\sum_{i,j = 1}^{n} a_i a_j f(x_i) k(x_i, x_j) f(x_j) = \sum_{i,j = 1}^{n} b_i k(x_i, x_j) b_j \geq 0, \quad \text{with } b_i = a_if(x_i)
$$

**(d)** Let $(k_m)$ be a series of p.d. kernels such that for every $x$, $y$, $\left(k_m(x, y)\right)_m$ converges. $$k(x, y) = \lim_{m \to +\infty} k_m(x, y)$$
Let $n \geq 1, x_1,...,x_n \in \mathcal{X}, \alpha \in \mathbb{R}^n$

We have  $\sum_{i,j = 1}^{n} a_i a_j k_m(x_i, x_j) \geq 0, \forall m$. Therefore, $\lim_{m\rightarrow\infty} \sum_{i,j = 1}^{n} a_i a_j k_m(x_i, x_j) \geq 0$. 
The sum being finite, we have 
$$\lim_{m\rightarrow\infty} \sum_{i,j = 1}^{n} a_i a_j k_m(x_i, x_j) = \sum_{i,j = 1}^{n} a_i a_j \lim_{m\rightarrow \infty}k(x_i, x_j) = \sum_{i,j = 1}^{n} a_i a_j k(x_i, x_j)  $$


**(e)** $k(x, y) = k_1(x, y) k_2(x, y)$ *This is accepted without proof*

**Corollary of (e)** $k(x, y) = (k_1(x, y))^d $ is pd for all $d \in \mathbb{N}$

### Implementation

Vector with vector:

$$k_\sigma(x, y) = \exp \left( -\frac{\|x-y\|^2}{2 \sigma^2}\right)$$


```python
def rbf_kernel_element_wise(x, y, sigma=1):
    '''
    returns the RBF (Gaussian) kernel k(x, y)
    
    Input:
    ------
    x and y are p-dimensional vectors 
    '''
    K = np.exp(-((x-y).dot(x-y))/(2*sigma**2))
    return K
```

Pairwise: return the matrix $\left[k_\sigma(X^1_i, X^2_j)\right]_{i,j}$

*Hint: expand $\|x_1-x_2\|^2$ again!*

```python
def rbf_kernel(X1, X2, sigma=10):
    '''
    Returns the kernel matrix K(X1_i, X2_j): size (n1, n2)
    
    Input:
    ------
    X1: an (n1, p) matrix
    X2: an (n2, p) matrix
    '''
    # For loop with rbf_kernel_element works but is slow in python
    # Use matrix operations!
    X2_norm = np.sum(X2 ** 2, axis = -1)
    X1_norm = np.sum(X1 ** 2, axis = -1)
    gamma = 1 / (2 * sigma ** 2)
    K = np.exp(- gamma * (X1_norm[:, None] + X2_norm[None, :] - 2 * np.dot(X1, X2.T)))
    return K
    return K
```

### Choosing $\sigma$
A practical approach of choosing $\sigma$ is called the "median heuristic":
    $$ \sigma \approx \mathrm{median} \left\{\|x_i-x_j\|:i,j=1,\dots, n\right\} \,.$$

Implement this heuristic:

```python
def sigma_from_median(X):
    '''
    Returns the median of ||Xi-Xj||
    
    Input
    -----
    X: (n, p) matrix
    '''
    euclidean_distances = np.sqrt(((X[None]-X[:,None])**2).sum(-1))
    return np.median(euclidean_distances)

print(sigma_from_median(X_train))
```

### Kernel Ridge Regression


#### General formulation

Let $\phi : \mathcal{X} \rightarrow \mathbb{R}^p$ be a feature mapping from the space of data to a Euclidean or Hilbert space. <br>
Let $f_\beta(x) = \beta^\top \phi(x)$ and $k: x,x' = \phi(x)^T\phi(x')$.
By the representer theorem, any solution $\beta$ of 
$$ \min_\beta  \frac 1n \sum_{i=1}^n (y_i - f_\beta(x_i))^2 + \lambda ||\beta||_2^2 $$
can be written as $$ \beta = \sum_i \alpha_i \phi(x_i) $$
and therefore,
$$ \hat{f}(x) = \sum_i \alpha_i \phi(x_i)^T \phi(x) = \sum_i \alpha_i k(x_i, x) $$
Let's define $K = [k(x_i, x_j)]_{i,j}$. Then
$$ \left( \hat{f}(x_1),..., \hat{f}(x_n) \right )^\top = K\alpha $$
and
$$ ||\beta||_2^2 = \sum_{i=1}^n\sum_{j=1}^n \alpha_i \alpha_j K(x_i, x_j) = \alpha^T K \alpha $$
and the problem can be rewritten as
 $$ \min_{\alpha \in \mathbb{R}^n}  \frac 1n (K\alpha - Y)^T(K\alpha - Y) + \lambda \alpha^T K \alpha $$
This problem has a unique solution:
$$ \alpha = (K + \lambda nI)^{-1} Y $$


#### Case $\phi(x) = x$

If $\phi(x) = x$, $K = [x_i^Tx_j]_{i,j} = XX^T$ and
$$ \beta = \sum_{i} \alpha_i x_i = X^T\alpha = X^T(XX^T + \lambda n I)^{-1}y $$

The prediction rule is 
$$\hat{y}(X_t) = X_t \beta = X_t X^\top(XX^\top + \lambda n I)^{-1} y$$

Starting from there, the 'kernel trick' can be applied to replace the pairwise dot products $XX^\top$ by a kernel matrix:
$$\color{green}{\hat{y}(X_t) = K_{X_t, X} (K_{X, X} + \lambda n I)^{-1} y}$$
